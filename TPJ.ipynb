```python
from networkx import grid_graph
import matplotlib.pyplot as plt
from networkx import nx
```

LATTICE

```python
G = grid_graph(dim=[2, 3, 4])
nx.draw(G,node_color='yellow', edge_color='green', with_labels=False)
```

This is to produce a 2 by 3 by 4 grid graph, a graph on 24 nodes

RANDOM ER

```python
n = 10  # 10 nodes
m = 20  # 20 edges

G = nx.gnm_random_graph(n, m)
nx.draw(G, node_color='white', edge_color='blue', with_labels=True)
```

This is to produce a random graph with 10 nodes and 20 edges

SMALL WORLD

```python
g = nx.watts_strogatz_graph(30,6,0.04, )

nx.draw(g, with_labels=True, node_color='pink',edge_color='grey')
```

Return a WattsStrogatz small-world graph with 30 nodes, 6 k nearest neighbors and a probability of 0.04
